```python
import numpy as np


A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])


result = np.dot(A, B)

print("Matrix A:\n", A)
print("Matrix B:\n", B)
print("Multiplication Result:\n", result)
```

```python
import numpy as np


A = np.array([[1, 2],
              [3, 4],
              [5, 6]])


A_transpose = A.T


B = np.array([[7, 8, 9],
              [10, 11, 12]])

dot_result = np.dot(A, B)

print("Matrix A (3x2):\n", A)
print("Transpose of A (2x3):\n", A_transpose)
print("Matrix B (2x3):\n", B)
print("Dot Product (A x B):\n", dot_result)
```
